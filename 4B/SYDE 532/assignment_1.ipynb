```python
import matplotlib.pyplot as plt
```

```python
def q(th, tc):
    return th - tc

def w(q, tc, th):
    return q * (1 - tc / th)

def th_prime(q):
    return -q

def tc_prime(q, w):
    return q - w

def step(th, tc, delta_t):
    q_ = q(th, tc)
    w_ = w(q_, tc, th)
    th_ = th_prime(q_)
    tc_ = tc_prime(q_, w_)
    return q_, w_, th + delta_t * th_, tc + delta_t * tc_
```

```python
delta_t = 0.01
th = 400
tc = 300

ws = []
qs = []
ths = []
tcs = []
for i in range(500):
    q_out, w_out, th, tc = step(th, tc, delta_t)
    qs.append(q_out)
    ws.append(w_out)
    ths.append(th)
    tcs.append(tc)

plt.plot(ths, label='Th(t)')
plt.plot(tcs, label='Tc(t)')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
plt.title('Temperature Over Time')
plt.legend()
plt.show()
```

```python
print("Final temperature: ", th, tc)
```

```python
efficiency = sum(ws) / sum(qs)
print("Simulated efficiency: ", efficiency)
```
